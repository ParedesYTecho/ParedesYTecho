#Ejercicio 2


```python
def process_text(text="Este es un texto de ejemplo.", words_to_replace=None):
    """
    Procesa una cadena de texto reemplazando ciertas palabras con asteriscos.

    Args:
    text (str): La cadena de texto a procesar.
    words_to_replace (list): Lista de palabras que serán reemplazadas por asteriscos.

    Returns:
    tuple: La cadena de texto procesada y la cantidad de palabras reemplazadas.
    """
    if words_to_replace is None:
        words_to_replace = ["ejemplo", "texto"]

    # Convertir la cadena a minúsculas y eliminar espacios en blanco iniciales y finales
    text = text.lower().strip()

    # Reemplazar palabras
    replaced_count = 0
    for word in words_to_replace:
        if word in text:
            text = text.replace(word, '*' * len(word))
            replaced_count += 1

    return text, replaced_count

def main():
    # Llamar a la función process_text
    processed_text, replaced_count = process_text()
    print(f"Texto procesado: {processed_text}")
    print(f"Cantidad de palabras reemplazadas: {replaced_count}")

# Llamar a la función principal
main()
```

#Ejercicio 3

```python
def is_prime(n):
    """
    Comprueba si un número n es primo.

    Args:
    n (int): El número a comprobar.

    Returns:
    bool: True si el número es primo, False en caso contrario.
    """
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

def prime_list(limit):
    """
    Genera una lista de números primos hasta un límite dado.

    Args:
    limit (int): El límite hasta el cual generar números primos.

    Returns:
    list: Lista de números primos hasta el límite dado.
    """
    primes = []
    for num in range(2, limit + 1):
        if is_prime(num):
            primes.append(num)
    return primes

def check_palindrome(primes):
    """
    Busca palíndromos en una lista de números primos.

    Args:
    primes (list): Lista de números primos.

    Returns:
    list: Lista de números primos que son palíndromos.
    """
    palindromes = []
    for prime in primes:
        if str(prime) == str(prime)[::-1]:
            palindromes.append(prime)
    return palindromes

def categorize_prime(prime):
    """
    Clasifica los números primos en "pequeños", "medianos" o "grandes".

    Args:
    prime (int): El número primo a clasificar.

    Returns:
    str: La categoría del número primo ("pequeño", "mediano" o "grande").
    """
    if prime < 10:
        return "pequeño"
    elif 10 <= prime < 100:
        return "mediano"
    else:
        return "grande"

def main():
    # Probar las funciones del programa
    limit = 150
    primes = prime_list(limit)

    print(f"Números primos hasta {limit}: {primes}")

    palindromes = check_palindrome(primes)

    print(f"Números primos que son palíndromos: {palindromes}")

    for prime in primes:
        category = categorize_prime(prime)
        print(f"El número primo {prime} es {category}.")

# Llamar a la función principal
main()
```

# Ejercicio 4


```python
def calculate_sum(numbers):
    """
    Calcula la suma de todos los números en la lista.

    Args:
    numbers (list): Lista de números enteros.

    Returns:
    int: La suma de todos los números.
    """
    return sum(numbers)

def calculate_positive_negative_sums(numbers):
    """
    Calcula la suma de los números positivos y la suma de los números negativos por separado.

    Args:
    numbers (list): Lista de números enteros.

    Returns:
    tuple: La suma de los números positivos y la suma de los números negativos.
    """
    positive_sum = sum(num for num in numbers if num > 0)
    negative_sum = sum(num for num in numbers if num < 0)
    return positive_sum, negative_sum

def find_max_min(numbers):
    """
    Encuentra el número máximo y el mínimo de la lista.

    Args:
    numbers (list): Lista de números enteros.

    Returns:
    tuple: El número máximo y el número mínimo.
    """
    return max(numbers), min(numbers)

def check_unique(numbers):
    """
    Comprueba si todos los números de la lista son únicos.

    Args:
    numbers (list): Lista de números enteros.

    Returns:
    bool: True si todos los números son únicos, False en caso contrario.
    """
    return len(numbers) == len(set(numbers))

def compare_positive_negative(numbers):
    """
    Utiliza el operador condicional para devolver un mensaje que indique si la lista contiene más números positivos o negativos.

    Args:
    numbers (list): Lista de números enteros.

    Returns:
    str: Mensaje indicando si hay más números positivos o negativos.
    """
    positive_count = sum(1 for num in numbers if num > 0)
    negative_count = sum(1 for num in numbers if num < 0)

    return "La lista contiene más números positivos." if positive_count > negative_count else "La lista contiene más números negativos."

def main():
    # Leer una lista de números enteros
    numbers = list(map(int, input("Ingrese una lista de números enteros separados por espacios: ").split()))

    while True:
        print("\nMenú:")
        print("1. Calcular la suma de todos los números.")
        print("2. Calcular la suma de los números positivos y la suma de los números negativos por separado.")
        print("3. Encontrar el número máximo y el mínimo de la lista.")
        print("4. Comprobar si todos los números de la lista son únicos.")
        print("5. Indicar si la lista contiene más números positivos o negativos.")
        print("6. Salir.")

        option = int(input("Seleccione una opción: "))

        if option == 1:
            total_sum = calculate_sum(numbers)
            print(f"La suma de todos los números es: {total_sum}")
        elif option == 2:
            positive_sum, negative_sum = calculate_positive_negative_sums(numbers)
            print(f"La suma de los números positivos es: {positive_sum}")
            print(f"La suma de los números negativos es: {negative_sum}")
        elif option == 3:
            max_num, min_num = find_max_min(numbers)
            print(f"El número máximo es: {max_num}")
            print(f"El número mínimo es: {min_num}")
        elif option == 4:
            unique = check_unique(numbers)
            print("Todos los números son únicos." if unique else "Hay números repetidos en la lista.")
        elif option == 5:
            message = compare_positive_negative(numbers)
            print(message)
        elif option == 6:
            break
        else:
            print("Opción no válida. Por favor, seleccione una opción del 1 al 6.")

# Llamar a la función principal
main()
```

#Ejercicio 5


```python
# Variable global para la lista de tareas
tasks = []

def add_task(task):
    """
    Agrega una nueva tarea a la lista global de tareas.

    Args:
    task (str): La tarea a agregar.
    """
    global tasks
    tasks.append(task)

def remove_task(task):
    """
    Elimina una tarea de la lista global de tareas, si existe.

    Args:
    task (str): La tarea a eliminar.
    """
    global tasks
    if task in tasks:
        tasks.remove(task)

def list_tasks():
    """
    Enumera todas las tareas almacenadas en la lista global de tareas.

    Returns:
    list: Lista de tareas.
    """
    global tasks
    return tasks

def task_manager():
    """
    Administra las operaciones de agregar, modificar y enumerar tareas.
    """
    def modify_task(new_task):
        """
        Modifica la primera tarea en la lista de tareas.

        Args:
        new_task (str): La nueva tarea que reemplazará la primera tarea.
        """
        global tasks
        if tasks:
            tasks[0] = new_task

    while True:
        print("\nMenú:")
        print("1. Agregar tarea.")
        print("2. Eliminar tarea.")
        print("3. Listar tareas.")
        print("4. Modificar primera tarea.")
        print("5. Salir.")

        option = int(input("Seleccione una opción: "))

        if option == 1:
            task = input("Ingrese la nueva tarea: ")
            add_task(task)
            print(f"Tarea '{task}' agregada.")
        elif option == 2:
            task = input("Ingrese la tarea a eliminar: ")
            remove_task(task)
            print(f"Tarea '{task}' eliminada.")
        elif option == 3:
            print("Lista de tareas:")
            for idx, task in enumerate(list_tasks(), start=1):
                print(f"{idx}. {task}")
        elif option == 4:
            new_task = input("Ingrese la nueva tarea para modificar la primera: ")
            modify_task(new_task)
            print(f"La primera tarea ha sido modificada a '{new_task}'.")
        elif option == 5:
            break
        else:
            print("Opción no válida. Por favor, seleccione una opción del 1 al 5.")

def main():
    # Llamar a la función task_manager para administrar las tareas
    task_manager()

# Llamar a la función principal
main()
```

#Ejercicio 6


```python
def print_strings_starting_with(strings, letter):
    """
    Imprime cada cadena que comience con una letra pedida al usuario.

    Args:
    strings (list): Lista de cadenas.
    letter (str): Letra con la que deben comenzar las cadenas.
    """
    for string in strings:
        if string.startswith(letter):
            print(string)

def count_strings_containing_substring(strings, substring):
    """
    Cuenta cuántas cadenas contienen una subcadena pedida al usuario.

    Args:
    strings (list): Lista de cadenas.
    substring (str): Subcadena a buscar en las cadenas.

    Returns:
    int: Cantidad de cadenas que contienen la subcadena.
    """
    count = 0
    for string in strings:
        if substring in string:
            count += 1
    return count

def find_longest_and_shortest(strings):
    """
    Encuentra la cadena más larga y la más corta de la lista.

    Args:
    strings (list): Lista de cadenas.

    Returns:
    tuple: La cadena más larga y la más corta.
    """
    if not strings:
        return None, None
    longest = max(strings, key=len)
    shortest = min(strings, key=len)
    return longest, shortest

def check_same_object(strings, index1, index2):
    """
    Verifica si dos cadenas específicas de la lista son en realidad el mismo objeto.

    Args:
    strings (list): Lista de cadenas.
    index1 (int): Índice de la primera cadena.
    index2 (int): Índice de la segunda cadena.

    Returns:
    bool: True si las dos cadenas son el mismo objeto, False en caso contrario.
    """
    return strings[index1] is strings[index2]

def check_length_greater_than_10(strings):
    """
    Verifica si alguna cadena de la lista tiene una longitud mayor que 10. Si no es así, imprime un mensaje que indique que ninguna cadena tiene más de 10 caracteres.

    Args:
    strings (list): Lista de cadenas.

    Returns:
    bool: True si alguna cadena tiene una longitud mayor que 10, False en caso contrario.
    """
    for string in strings:
        if len(string) > 10:
            return True
    else:
        print("Ninguna cadena tiene más de 10 caracteres.")
        return False

def main():
    # Leer una lista de cadenas desde teclado
    strings = input("Ingrese una lista de cadenas separadas por comas: ").split(',')

    while True:
        print("\nMenú:")
        print("1. Imprimir cada cadena que comience con una letra pedida al usuario.")
        print("2. Contar cuántas cadenas contienen una subcadena pedida al usuario.")
        print("3. Encontrar la cadena más larga y la más corta de la lista.")
        print("4. Verificar si dos cadenas específicas de la lista son en realidad el mismo objeto.")
        print("5. Verificar si alguna cadena de la lista tiene una longitud mayor que 10.")
        print("6. Salir.")

        option = int(input("Seleccione una opción: "))

        if option == 1:
            letter = input("Ingrese la letra: ")
            print_strings_starting_with(strings, letter)
        elif option == 2:
            substring = input("Ingrese la subcadena: ")
            count = count_strings_containing_substring(strings, substring)
            print(f"Cantidad de cadenas que contienen la subcadena '{substring}': {count}")
        elif option == 3:
            longest, shortest = find_longest_and_shortest(strings)
            print(f"La cadena más larga es: '{longest}'")
            print(f"La cadena más corta es: '{shortest}'")
        elif option == 4:
            index1 = int(input("Ingrese el índice de la primera cadena: "))
            index2 = int(input("Ingrese el índice de la segunda cadena: "))
            same_object = check_same_object(strings, index1, index2)
            print(f"Las cadenas en los índices {index1} y {index2} {'son' if same_object else 'no son'} el mismo objeto.")
        elif option == 5:
            check_length_greater_than_10(strings)
        elif option == 6:
            break
        else:
            print("Opción no válida. Por favor, seleccione una opción del 1 al 6.")

# Llamar a la función principal
main()
```
